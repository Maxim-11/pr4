```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv('vgsales_4.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
dfr = df.drop(columns=['NA_Sales'])
dfr
```

```python
dfg = df.replace({'Publisher' : np.nan}, '1111')
dfg.isnull().sum()
```

```python
simple = SimpleImputer(strategy='mean', missing_values=np.nan)
simple = simple.fit(df[['Year']])
df['Year'] = simple.transform(df[['Year']])
df.isnull().sum()
```

```python
df.sort_values(by='Year', ascending=True)
```

```python
df.to_csv('df.csv')
dfr.to_csv('dfr.csv')
dfg.to_csv('dfg.csv')

```
